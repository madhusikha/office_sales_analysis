```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,5),dpi=200)  
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education')

# since 'level of education' is a categorical value, it showed different colors among salaries & sales-categorization
```

```python
plt.figure(figsize=(10,5),dpi=200)  
sns.scatterplot(x='salary',y='sales',data=df,hue='salary')

# since 'salary' is a continuous value, the color is a gradient
```

```python
plt.figure(figsize=(10,5),dpi=200)  
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience')

# since 'salary' is a continuous value, the color is a gradient
```

```python
plt.figure(figsize=(10,5),dpi=200)  

sns.scatterplot(x='salary',y='sales',data=df,style='level of education',
                hue='level of education',palette='Dark2')

# 'style'--marker style
# we have to give categorical column only
```

```python
plt.figure(figsize=(10,5),dpi=200)  

sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2',size='salary')

# size of the dots is varied as per 'salary'
```

```python
plt.figure(figsize=(10,5),dpi=200)  

sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2',size='work experience')

# size varied w.r.t. 'work experience' (categorical)
```

# Distribution plots

```python
plt.figure(figsize=(5,3),dpi=150)  # this line is not mandatory. if we wanna increase size of fig window use this
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
sns.displot(data=df,x='salary',bins=60)  # distribution of 'salary'
```

```python

sns.histplot(data=df,x='salary')
```

```python
# we can plot ONLY 'kdeplot'

sns.kdeplot(data=df,x='salary',shade=True)
```

# Categorical Plots- Statistics within Categories 

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=150)
sns.countplot(data=df,x='division')
```

```python
plt.figure(figsize=(10,4),dpi=150)
sns.countplot(data=df,x='level of education')
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=150)
sns.countplot(data=df,x='level of education',hue='division')
```

```python
plt.figure(figsize=(10,4),dpi=150)
sns.barplot(data=df)

# each bar in the below plot is the average salary in that category
# the black vertical line is the Standard Deviation-'sd'. which we set as " ci='sd' "
```

```python
plt.figure(figsize=(10,4),dpi=150)
sns.barplot(data=df,x='level of education',y='salary',estimator=np.mean,ci='sd',hue='division')
```

```python
plt.figure(figsize=(10,5),dpi=200)
sns.boxplot(data=df,y='sales',x='level of education',hue='division')


# we can see distributions of continuous features and their relation to different categories

# BOX PLOT:

# plot a continuous feature against categorical feature(s). If we want sub-categories, use hue option

# continuous features: [math, reading and writing scores]

# Categorical features: [gender, race, parent level of education, lunch, test preparation course]

```

```python
plt.figure(figsize=(10,5),dpi=200)
sns.boxplot(data=df,y='sales',x='work experience')
```
